```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def recursion(n):
    if n == 1:
        return 1
    elif n == 2:
        return 1/2
    else:
        return 1/2 * recursion(n - 1) + 1/2 * recursion(n - 2)

recursion(25)
```

```python
def explicit_recursion(n):
    return -2/3 * (-1/2) ** n + 2/3

explicit_recursion(25)
```

```python
def monte_carlo_prob(trials, target):
    hits = 0  
    for _ in range(trials):
        position = 1  
        for _ in range(24):  
            coin = random.choice([1, 2])  
            position += coin 
            if position == target:  
                hits += 1  
    return hits / trials

monte_carlo_prob(100000, 25)
```

```python
n_val = range(1, 50)
p = [explicit_recursion(n) for n in n_val]

plt.figure(figsize = [10, 8])
plt.title('Graph of P(n) against n')
plt.plot(n_val, p)
plt.xlabel('n')
plt.ylabel('P(n)')
plt.grid(True)
plt.show()
```

```python
def general_explicit_recursion(n, p):
    return (1 - p) / ((2 - p) * (p - 1)) * (p - 1) ** n + 1 / ( 2 - p)

general_explicit_recursion(25, 0.5)
```

```python
p_val = np.linspace(0, 1, 100)
P = [general_explicit_recursion(25, p) for p in p_val]

plt.figure(figsize = [10, 8])
plt.title('Graph of P(25) against p')
plt.plot(p_val, P)
plt.xlabel('p')
plt.ylabel('P(25, p)')
plt.grid(True)
plt.show()
```
